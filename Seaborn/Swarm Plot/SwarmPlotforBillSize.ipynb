program to draw a swarm plot of total bill against size for a given dataset

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-22 10:34
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : SwarmPlotforBillSize  
'''
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

try:
    df = pd.read_csv('D:\Python\Jupyter Notebook\Seaborn\Scatterplot\data.csv')

    # Incresing size of the plot BCoz all points are not printing(Showing warning)
    plt.figure(figsize=(10,8))
    sns.swarmplot(x = "size", y = "total_bill", data = df)
    plt.title("SwarmPlot for Total_Bill.")
    plt.show()
    
except Exception as e:
    print(e)
```
