program to draw a point plot for sex against survived for a dataset given in url

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-21 09:48
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : Pointplot  
'''
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

try:
    df = pd.read_csv('D:\Python\Jupyter Notebook\Seaborn\Barchart\\titanic.csv')

    sns.pointplot(x = "sex", y = "survived",data = df, markers='o',color='r',linestyles='-.' )
    plt.title('Point Plot for Sex Ratio')
    plt.show()
    
except Exception as e:
    print(e)
```
