program to draw a violin plot of sex against day for a given dataset

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-21 10:09
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : ViolinPlot  
'''
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

try:
    df = pd.read_csv('D:\Python\Jupyter Notebook\Seaborn\Scatterplot\data.csv')

    plt.figure(figsize=(10,8))

    sns.violinplot(x = "tip", y = "day", data = df)
    plt.title("ViolinPlot for Tips of Days")
    plt.show()
    
except Exception as e:
    print(e)
```
