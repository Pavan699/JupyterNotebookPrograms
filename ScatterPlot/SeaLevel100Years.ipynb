Python program to draw a scatter plot to find sea level rise in past 100 years.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-21 03:56
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : SeaLevel100years  
'''
from matplotlib import pyplot as plt
import pandas as pd

try:
    data = pd.read_csv('sealevel.csv')

    time = data['Time']
    sea_level = data['GMSL']

    plt.scatter(time, sea_level, edgecolors='g')

    plt.xlabel('Year' ,fontsize = 15)  
    plt.ylabel('Sea Level (inches)',fontsize = 15)
    plt.title('Rise in Sea level',fontsize = 15)
    
    plt.show()
    
except Exception as e:
    print(e)
```
