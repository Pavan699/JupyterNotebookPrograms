Python program to draw a scatter plot for three different groups camparing weights and heights.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-21 03:14
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : CompareWeightHeight  
'''
from matplotlib import pyplot as plt
import numpy as np

try:
    weight1=[67,57.2,59.6,70.64,55.8,61.2,60.45,61,56.23,56]
    height1=[101.7,197.6,98.3,125.1,113.7,157.7,136,148.9,125.3,114.9] 
    
    weight2=[61.9,64,62.1,55.2,62.3,65.4,62.4,61.4,70.5,63.6]
    height2=[152.8,155.3,135.1,125.2,151.3,135,182.2,195.9,165.1,125.1] 
    
    weight3=[68.2,67.2,58.4,68.7,62,71.3,70.8,70,71.1,71.7]
    height3=[165.8,170.9,192.8,135.4,161.4,136.1,167.1,235.1,181.1,177.3]
     
    plt.scatter(weight1, height1, color='green',edgecolors='k')
    plt.scatter(weight2, height2, color='red',edgecolors='k')
    plt.scatter(weight3, height3, color='yellow',edgecolors='k')
    plt.xlabel('weight in Kg')
    plt.ylabel('height in cm')
    plt.title('Weight vs Height')
    plt.legend([''])
    plt.show()
    
except Exception as e:
    print(e)
```
