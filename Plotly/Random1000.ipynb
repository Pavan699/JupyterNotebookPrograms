program to draw a scatter plot for random 1000 x and y coordinates

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-22 01:03
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : ScatterPlotfor1000points  
'''
import plotly.express as px
from pylab import randn
import pandas as pd

try:
    x = randn(1000)
    y = randn(1000)
    df = pd.DataFrame({ 'x-axis':x, 'y-axis':y })

    fig = px.scatter(df, x='x-axis', y='y-axis', width=700, height=500, title="Scatter Plot")
    fig.show()
    
except Exception as e:
    print(e)
```
