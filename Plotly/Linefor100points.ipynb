Program to draw line and scatter plots for random 100 x and y coordinates

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-22 01:27
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : Linefor100points  
'''
import plotly.express as px
from pylab import randn
import pandas as pd

try:
    x = randn(100)
    y = randn(100)
    df = pd.DataFrame({ 'x-axis':x, 'y-axis':y })

    fig = px.line(df, x='x-axis', y='y-axis', width=700, height=500, title="Line Plot")
    fig.show()
    
except Exception as e:
    print(e)
```

```python
fig = px.scatter(df, x='x-axis', y='y-axis', width=700, height=500, title="Scatter Plot")
fig.show()
```
