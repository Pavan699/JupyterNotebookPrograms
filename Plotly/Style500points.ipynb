program to draw a scatter plot for random 500 x and y coordinates and style it

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-22 01:49
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : Style500points  
'''
import plotly.express as px
from pylab import randn
import pandas as pd

try:
    x = randn(500)
    y = randn(500)
    df = pd.DataFrame({ 'x-axis':x, 'y-axis':y })

    fig = px.scatter(df, x='x-axis', y='y-axis', width=800, height=600, title="Scatter Plot",color="y-axis")
    fig.show()
    
except Exception as e:
    print(e)
```
