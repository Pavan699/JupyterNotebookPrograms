Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-19 11:52
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : FinancialDataChart  
'''
from matplotlib import pyplot as plt
import pandas as pd

try:
    data = pd.read_csv("fdata.csv", sep=',', parse_dates=True, index_col=0)
    data.plot()
    plt.title("Financial data of Alphabet Inc.")
    plt.ylabel("Alphabet Inc.")
    plt.show()
    
except Exception as e:
    print(e)
```
