Python program to plot two or more lines on same plot with suitable legends of each line.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-19 02:44
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : TwoLinesOnPlot  
'''
from matplotlib import pyplot as plt

try:
    x = [1,2,3,4,5,6]
    y = [num**3 for num in x]
    print(x)
    print(y)

    plt.plot(x, label = "Numbers")
    plt.plot(y, label = "Cube of number")

    plt.xlabel("Numbers")
    plt.ylabel("Cubes of Numbers")
    plt.title("Two lines on same plot with legend")

    plt.legend()
    plt.show()
    
except Exception as e:
    print(e)
```
