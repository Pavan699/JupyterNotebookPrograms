Python program to plot two or more lines and set the line markers.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-19 03:12
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : LineMarkers  
'''
from matplotlib import pyplot as plt

try:
    x = [1,3,5,7,9]
    y = [num+2 for num in x]
    print(x)
    print(y)

    plt.plot(x, label = "Numbers",marker='.')
    plt.plot(y, label = "After adding 2",marker='o')

    plt.xlabel("Numbers")
    plt.ylabel("Numbers with addition")
    plt.title("Adding 2 in given numbers")

    plt.legend()
    plt.show()
    
except Exception as e:
    print(e)
```
