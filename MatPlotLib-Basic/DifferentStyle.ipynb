Python program to plot two or more lines with different styles

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-19 03:04
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : DiffrentStyle  
'''
from matplotlib import pyplot as plt

try:
    x = [1,2,3,4,5,6]
    y = [num**3 for num in x]
    print(x)
    print(y)

    plt.plot(x, label = "Numbers",linestyle='--')
    plt.plot(y, label = "Cube of number",linestyle='-.')

    plt.xlabel("Numbers")
    plt.ylabel("Cubes of Numbers")
    plt.title("Two lines with different styles")

    plt.legend()
    plt.show()
    
except Exception as e:
    print(e)
```
