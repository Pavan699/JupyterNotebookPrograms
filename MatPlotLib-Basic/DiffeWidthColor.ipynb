Python program to plot two or more lines with legends, different widths and colors.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-19 02:52
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : DiffeWidthColor  
'''
from matplotlib import pyplot as plt

try:
    x = [1,2,3,4,5,6]
    z = [s**2 for s in x]
    y = [s**3 for s in x]
    print(x)
    print(z)
    print(y)

    plt.plot(x, y,label = "Cube of numbers",color='b',linestyle='--',marker='o',linewidth=3)
    plt.plot(x, z, label = "Squre of number",color='r',marker='o',linewidth=3)

    plt.xlabel("Numbers")
    plt.ylabel("Power of Numbers")
    plt.title("Two lines on same plot with legend")

    plt.legend()
    plt.show()
    
except Exception as e:
    print(e)
```
