Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-20 12:11
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : CSVFilePiechart  
'''
from matplotlib import pyplot as plt
import pandas as pd

try:
    df =  pd.read_csv('model.csv')
    #seprating the data 
    country_data = df["country"] # names 
    medal_data = df["gold_medal"] # medal count

    explode = (0.1, 0, 0, 0, 0)  

    plt.pie(medal_data, labels=country_data, explode=explode,startangle=90)

    plt.title("Country with their total gold medals")
    plt.show()
    
except Exception as e:
    print(e)
```
