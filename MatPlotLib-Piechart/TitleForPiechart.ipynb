Python programming to create a pie chart with a title of the popularity of programming Languages.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-20 11:20
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : TitleForPiechart  
'''
from matplotlib import pyplot as plt

try:
    languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
    popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        
    plt.pie(popuratity, labels=languages,startangle=90) # adding startangle
    plt.title("Popularity of Programming Languages") # adding title
    plt.show()
    
except Exception as e:
    print(e)
```
