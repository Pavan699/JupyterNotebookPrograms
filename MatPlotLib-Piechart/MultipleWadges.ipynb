Python programming to create a pie chart with a title of the popularity of programming Languages.Make multiple wedges of the pie.

```python
'''
@Author: Pavan Nakate
@Date: 2021-11-20 11:50
@Last Modified by: Pavan Nakate
@Last Modified time: None
@Title : MultipleWadges  
'''
from matplotlib import pyplot as plt

try:
    languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
    popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    # Adding colors
    color = ['r','g','b','y','orange','gray']
    explode = [0,0,0,0.2,0,0]
        
    plt.pie(popuratity, labels=languages,startangle=90,colors=color,wedgeprops={'edgecolor':'k'},explode=explode) 
    plt.title("Popularity of Programming Languages") # adding title
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(e)
```
